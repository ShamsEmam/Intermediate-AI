```python
def outliears(col_name):
    global New_df ,Q3,Q1,IQR
    Q1=New_df[col_name].quantile(0.25)
    Q3=New_df[col_name].quantile(0.75)
    IQR=Q3-Q1
    min_data=Q3+1.5*IQR
    max_data=Q3-1.5*IQR
    
    return min_data ,max_data

min_data ,max_data=outliears('PetalLengthCm')
```

# K_Means

```python
from sklearn.cluster import KMeans
```

```python
K_range =range(1,11)
sse=[]
for i in K_range:
    kmeans=KMeans(n_clusters=i ,random_state=42)
    kmeans.fit_predict(data)
    sse.append(kmeans.inertia_)
```

```python
sse
```

```python
plt.plot(K_range,sse)
plt.plot(range(1,11),sse,linewidth=4,markersize=12,marker='o')
plt.xlabel('Number of K')
plt.ylabel('SSE')
plt.show()
```
